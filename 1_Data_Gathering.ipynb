# **1.Working With CSV Files**

**LOCAL FILE**

```python
#When the file is present locally
import pandas as pd
df=pd.read_csv("diabetes.csv")
df.head(10)
```

**Opening CSV Files From Links**

```python
#opening csv from url
import requests
from io import StringIO
url="https://raw.githubusercontent.com/datasets/gdp/main/data/gdp.csv" # Replace with your actual CSV URL
headers= {} #heere we giive the link headers
req=requests.get(url,headers=headers)
df=pd.read_csv(StringIO(req.text))
df.head(10)
```

**SEP Parameter**

```python
pd.read_csv("diabetes.csv", sep='\t')#It helps to give the spaces between the columns
```

**Index_col Parameter**

```python
pd.read_csv("diabetes.csv",index_col='Glucose')
```

**Header**

```python
pd.read_csv("diabetes.csv",header=0)
```

**Use COl parameter**

```python
pd.read_csv("diabetes.csv",usecols=['Glucose','Age', 'BMI'])
```

**Squeeze**

```python
# The 'squeeze' parameter is deprecated and removed in recent pandas versions.
# To get a Series from a single column, read the CSV and then select the column.
pd.read_csv("diabetes.csv", usecols=['Glucose'])['Glucose']
```

**SkipRows / Rows Prameter**

```python
pd.read_csv("diabetes.csv",skiprows=[1,3])
```

**nRows**

```python
#give the  number of rows as requested(n)
pd.read_csv("diabetes.csv",nrows=100)
```

**Encoding Parameter**

```python
#used when the text isside isn't encoded... like if it's a emoji dataset
pd.read_csv("diabetes.csv",encoding='latin-1')
```

**Skip Bad Lines**

```python
pd.read_csv("diabetes.csv",sep='1',encoding="latin-1",on_bad_lines='skip')
```

**dTypes Parameter**

```python
#help to change a datatype of a specified column
pd.read_csv("diabetes.csv",dtype={'Glucose':'float64'})
```

**Handling Dates**

```python
import pandas as pd
pd.read_csv("IPL.csv",parse_dates=['date']).info()
```

**Converters**

```python
from os import rename
pd.read_csv("IPL.csv",converters={'bowling_team':rename})
```

**Na_Values Parameter**

```python
pd.read_csv("IPL.csv",na_values=['Mi'])
```

**Loading A huge Dataset in chunks**

```python
dfs=pd.read_csv("IPL.csv",chunksize=10000)
for chunks in dfs:
  print(chunks.shape)
```
